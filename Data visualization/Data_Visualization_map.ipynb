```python
import pandas as pd
forbes_data = pd.read_csv('forbes_top_2000_world_largest_public_companies.csv')
country_counts = forbes_data.groupby('contry/territory').size().reset_index(name='counts')
```

```python
import geopandas as gpd

# Load world map
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

```

```python
# Merge data
merged = world.set_index('name').join(country_counts.set_index('contry/territory'))
```

```python
import folium

# Create map object
m = folium.Map(location=[20, 0], zoom_start=2)

# Add the choropleth layer
folium.Choropleth(
    geo_data=world,
    name='choropleth',
    data=country_counts,
    columns=['contry/territory', 'counts'],
    key_on='feature.properties.name',
    fill_color='YlGn',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Number of Companies'
).add_to(m)

# Display the map
m.save('map.html')
```
